```
%matplotlib inline
import cv2
```

```
img = cv2.imread('sample.png')
print(type(img))
print(img.shape)
print(img.dtype)
```

```
import matplotlib.pyplot as plt
```

```
plt.imshow(img)
```

```
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```
from PIL import Image
img = Image.open('sample.png')
print(img)
img
```

```
import numpy as np
img = np.array(img)
print(type(img))
print(img.shape)
print(img.dtype)
```

```
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(img_gray)
```

```
plt.gray()
plt.imshow(img_gray)
```

```
img_gray.shape
```

```
kernel = np.array([[-1,0,1],
                   [-1,0,1],
                   [-1,0,1]
                   ,])
kernel.shape
```

```
img_conv = cv2.filter2D(img_gray, 0, kernel)
plt.gray()
plt.imshow(img_conv)
```

```
kernel = np.array([
    [-1,-1,-1],
    [0,0,0],
    [1,1,1]
])
kernel.shape
```

```
img_conv = cv2.filter2D(img_gray, -1, kernel)
plt.gray()
plt.imshow(img_conv)
```
